```python
import bs4
import requests
import pandas as pd
```

```python
link="https://www.glassdoor.co.in/Job/india-data-scientist-jobs-SRCH_IL.0,5_IN115_KO6,20_IP3.htm?includeNoSalaryJobs=true&pgc=AB4AAoEAPAAAAAAAAAAAAAAAAd5581YAbAECAQgkDgHxkoSzPmNguLpLp5LxaU0Pg5bQubHA4Q743xVBeaP1azTLsm4rZXJy9uvaFoBpnQ7%2B14uCmjWObfYfEbhoiJJDlmXDe14ziffxG9Kin2I6%2FfS2SM5Jf2iUJaZ8YTqv2iE1K3QQXAAA"
```

```python
req=requests.get(link)
```

```python
soup=bs4.BeautifulSoup(req.content,"html.parser")
```

```python
sal=[]
for i in soup.findAll('span',{'class':'job-search-key-1hbqxax e1wijj240'}):
    sal.append(i.text)
```

```python
name=[]
nm=soup.find_all('div',{'class':'d-flex justify-content-between align-items-start'})
for i in range(len(sal)):
    name.append(nm[i].text)
```

```python
locn=[]
lc=soup.find_all('span',{'class':'css-1buaf54 pr-xxsm job-search-key-iii9i8 e1rrn5ka4'})
for i in range(len(sal)):
    locn.append(lc[i].text)
```

```python
ds=[]
desc= soup.find_all('a',{'class':'jobLink job-search-key-1rd3saf eigr9kq1'})
for i in range(len(sal)):
    ds.append(desc[i].text)
```

```python
data={'Company Name':name,'Location':locn,"Job Description":ds,'Salary Estimate':sal }
```

```python
df1=pd.DataFrame(data)
```

```python
df1
```

```python
salary=df1['Salary Estimate'].apply(lambda x: x.split('(')[0])
```

```python
sign=salary.apply(lambda x: x.replace('₹',''))
```

```python
var=sign.apply(lambda x: x if '-' in x else '0-0')
```

```python
df1['min_salary']=var.apply(lambda x: x.split('-')[0].strip())
```

```python
df1['min_salary']=df1.min_salary.apply(lambda x: int(x.replace(',','')))
```

```python
df1['max_salary']=var.apply(lambda x: x.split('-')[1].strip())
```

```python
df1['max_salary']=df1.max_salary.apply(lambda x: int(x.replace(',','')))
```

```python
df1
```

```python
df1['avg_salary']=(df1.min_salary+df1.max_salary)/2
```
