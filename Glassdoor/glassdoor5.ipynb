```python
link="https://www.glassdoor.co.in/Job/india-data-science-jobs-SRCH_IL.0,5_IN115_KO6,18_IP5.htm?includeNoSalaryJobs=true&pgc=AB4ABIEAeAAAAAAAAAAAAAAAAd6tmq0AxQEEARQqDCYLIAcv%2FPwkPoxQ2jrZ9Ca4OfHvnER4cp10WJVTNuv98592P3xXpWJ67j2TdyoVh%2FMYLRL4Pqbsa9vmZObXGTw47YdTVfNvdGC7quLs6N4d2W6gINgjUK6T6QQ0%2B7nU7UUI3TL6x6xOY7s0b99R4a89Wd051InLVi8mKR9vriP%2F84eO%2BpUpynFmyftVdkZGxKTaL0ljR2%2FvK573MpUbfvGYSpEmK355x5VvNRVoIS7WRROr97d%2BRFvs9243QnsiAAA%3D"
```

```python
req=requests.get(link)
```

```python
soup=bs4.BeautifulSoup(req.content,'html.parser')
print(soup.prettify())
```

```python
sal=[]
for i in soup.findAll('span',{'class':'job-search-key-1hbqxax e1wijj240'}):
    sal.append(i.text)
```

```python
name=[]
nm=soup.find_all('div',{'class':'d-flex justify-content-between align-items-start'})
for i in range(len(sal)):
    name.append(nm[i].text)
```

```python
locn=[]
lc=soup.find_all('span',{'class':'css-1buaf54 pr-xxsm job-search-key-iii9i8 e1rrn5ka4'})
for i in range(len(sal)):
    locn.append(lc[i].text)
```

```python
ds=[]
desc= soup.find_all('a',{'class':'jobLink job-search-key-1rd3saf eigr9kq1'})
for i in range(len(sal)):
    ds.append(desc[i].text)
```

```python
data={'Company Name':name,'Location':locn,"Job Description":ds,'Salary Estimate':sal}
```

```python
df2=pd.DataFrame(data)
```

```python
df2
```

```python
salary=df2['Salary Estimate'].apply(lambda x: x.split('(')[0])
```

```python
sign=salary.apply(lambda x: x.replace('',''))
```

```python
var=sign.apply(lambda x: x if '-' in x else '0-0')
```

```python
df2['min_salary']=var.apply(lambda x: x.split('-')[0].strip())
```

```python
df2['min_salary']=df2.min_salary.apply(lambda x: int(x.replace(',','')))
```

```python
df2['max_salary']=var.apply(lambda x: x.split('-')[1].strip())
```

```python
df2['max_salary']=df2.max_salary.apply(lambda x: int(x.replace(',','')))
```

```python
df2.head()
```

```python
df2['avg_salary']=(df2.min_salary+df2.max_salary)/2
```
