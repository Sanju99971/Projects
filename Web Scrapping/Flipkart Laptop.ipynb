```python
import bs4
import requests 
import pandas as pd
```

```python
link='https://www.flipkart.com/search?q=laptops&otracker=search&otracker1=search&marketplace=FLIPKART&as-show=on&as=off'
```

```python
data=requests.get(link)
data
```

```python
data.text
```

```python
soup=bs4.BeautifulSoup(data.text,'html.parser')
print(soup.prettify())
```

```python
Laptop=soup.find_all("div",{'class':'_4rR01T'})
Laptop
```

```python
laptop=[]
for laptops in Laptop:
    lap=laptops.get_text()
    laptop.append(lap)
laptop
print(len(laptop))   
```

```python
Rating=soup.find_all('div',{"class":'_3LWZlK'})
Rating
```

```python
Stars=[]
for stars in Rating:
    stars=stars.get_text().replace(',',"")
    stars=stars[0:24]
    Stars.append(stars)
Stars
    
    
```

```python
price=soup.find_all('div',{'class':'_30jeq3 _1_WHN1'})
price
```

```python
Price=[]
for cash in price:
    cash=cash.get_text()[1:]
    Price.append(cash)
Price
```

```python
discount=soup.find_all('div',{'class':"_3Ay6Sb"})
discount
```

```python
Discount=[]
for dis in discount:
    dis=dis.get_text().replace("off","")
    Discount.append(dis)
Discount
```

```python
Data=pd.DataFrame()
Data["Laptop"]=laptop
Data["Rating"]=Stars
Data["Price"]=Price
Data["Discount"]=Discount
```

```python
Data.to_csv("Flipkart.csv", index=False)
```
