```python
from bs4 import BeautifulSoup as bs
import requests 
import pandas as pd
import re
```

```python
link='https://www.bookdepository.com/category/1897/Computing'
```

```python
response=requests.get(link)
response
```

```python
soup=bs(response.text,"html.parser")
print(soup.prettify())
```

```python
Books=[]
for i in soup.find_all('h3',{'class':"title"}):
    book=i.get_text(strip=True)
    Books.append(book)
```

```python
df1=pd.DataFrame(Books)
```

```python
Author=[]
for i in soup.find_all('p',{'class':'author'} ):
    author=i.get_text(strip=True)
    Author.append(author)
```

```python
df2=pd.DataFrame(Author)
```

```python
Price=[]
for i in soup.find_all('span',{"class":'sale-price'}):
    price=i.get_text()
    Price.append(price)
```

```python
df3=pd.DataFrame(Price)
```
