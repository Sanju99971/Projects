```python
import bs4
import requests
import pandas as pd
```

```python
link='https://www.flipkart.com/search?q=Mobiles&otracker=search&otracker1=search&marketplace=FLIPKART&as-show=on&as=off'
```

```python
data=requests.get(link)
data
```

```python
data.text
```

```python
soup=bs4.BeautifulSoup(data.text ,'html.parser')
print(soup.prettify())
```

```python
phone=soup.find_all('div',{'class':"_4rR01T"})
phone
```

```python
Phones=[]
for phon in phone:
    phon=phon.get_text()
    Phones.append(phon)
Phones
A=pd.DataFrame(Phones)
A
```

```python
rate=soup.find_all("div",{'class':'_30jeq3 _1_WHN1'})
rate
Price=[]
for Rate in rate:
    Rate=Rate.get_text()[1:]
    Price.append(Rate)
Price
B=pd.DataFrame(Price)
B
```

```python
discount=soup.find_all('div',{'class':"_3Ay6Sb"})

discount
```

```python
Discount=[]
for Dis in discount:
    Dis=Dis.get_text().replace("off","")
    
    Discount.append(Dis)
Discount
```

```python
Discount=Discount[0:24]
```

```python
C=pd.(Discount)
C
```

```python
Data=pd.DataFrame()
Data['Phones']=Phones
Data["Price"]=Price
Data["Discount"]=Discount
```

```python
df=Data.to_csv('Phone.csv', index=False)
```
